# Loading Dataset

```python
import pandas as pd
cases = pd.read_csv("owid-covid-data.csv")
cases.head()
```

# Handling Missing Values

```python
# Calculating percentage of null values in each column and keeping track of them in a separate dictionary
column_list = list(cases.columns)
mydictionary = {}
print("Columns :\n")
for x in column_list:
    calc = cases[x].isna().sum()/cases.shape[0]*100
    #print("{} - ".format(x),calc)
    mydictionary[x] = calc
    
for key in mydictionary:
    print("{} - {}".format(key,mydictionary[key]))
```

```python
# Dropping entire columns with more than 50% null values and updating our "cases" dataframe

columns_to_be_dropped = []
for key in mydictionary:
    if mydictionary[key] > 50:
        columns_to_be_dropped.append(key)
        
print(columns_to_be_dropped)
```

So there is no column in our dataframe which has more than 50% null values. Thus no need to drop any column.

```python
# Before modifying the null values
cases.info()
```

```python
# Replacing the null values in each continuous numerical column by mean of that column

cases["new_cases_smoothed"].fillna(cases["new_cases_smoothed"].mean(),inplace=True)
cases["new_deaths_smoothed"].fillna(cases["new_deaths_smoothed"].mean(),inplace=True)
cases["total_cases_per_million"].fillna(cases["total_cases_per_million"].mean(),inplace=True)
cases["new_cases_per_million"].fillna(cases["new_cases_per_million"].mean(),inplace=True)
cases["new_cases_smoothed_per_million"].fillna(cases["new_cases_smoothed_per_million"].mean(),inplace=True)
cases["total_deaths_per_million"].fillna(cases["total_deaths_per_million"].mean(),inplace=True)
cases["new_deaths_per_million"].fillna(cases["new_deaths_per_million"].mean(),inplace=True)
cases["new_deaths_smoothed_per_million"].fillna(cases["new_deaths_smoothed_per_million"].mean(),inplace=True)
cases["total_tests_per_thousand"].fillna(cases["total_tests_per_thousand"].mean(),inplace=True)
cases["new_tests_per_thousand"].fillna(cases["new_tests_per_thousand"].mean(),inplace=True)
cases["new_tests_smoothed_per_thousand"].fillna(cases["new_tests_smoothed_per_thousand"].mean(),inplace=True)
cases["tests_per_case"].fillna(cases["tests_per_case"].mean(),inplace=True)
cases["positive_rate"].fillna(cases["positive_rate"].mean(),inplace=True)
cases["stringency_index"].fillna(cases["stringency_index"].mean(),inplace=True)



# Replacing the null values in each ordinal numerical column by mode of that column

cases["total_cases"].fillna(cases["total_cases"].mode()[0],inplace=True)
cases["new_cases"].fillna(cases["new_cases"].mode()[0],inplace=True)
cases["total_deaths"].fillna(cases["total_deaths"].mode()[0],inplace=True)
cases["new_deaths"].fillna(cases["new_deaths"].mode()[0],inplace=True)
cases["new_tests"].fillna(cases["new_tests"].mode()[0],inplace=True)
cases["total_tests"].fillna(cases["total_tests"].mode()[0],inplace=True)
cases["new_tests_smoothed"].fillna(cases["new_tests_smoothed"].mode()[0],inplace=True)
```

# Converting date column to ordinal

```python
# Converting date column to ordinal
import datetime as dt
cases["date"]=pd.to_datetime(cases["date"])
cases["date"]=cases["date"].map(dt.datetime.toordinal)
```

```python
cases.info()
```

# Dropping Useless Columns

```python
# Dropping useless columns
cases1 = cases
cases = cases.drop(["iso_code","continent","location","gdp_per_capita","extreme_poverty","cardiovasc_death_rate","diabetes_prevalence","female_smokers","male_smokers"],axis=1)
```

```python
cases.info()
```

# Converting useful categorical columns to numerical

```python
# Converting useful categorical columns to numerical by labelencoder

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
labelencoder.fit(cases["tests_units"])
cases["tests_units"] = labelencoder.transform(cases["tests_units"])
```

```python
cases.info()
```

```python
cases.head()
```

# Selecting Features and Target Variable

```python
Y = cases["total_cases"]
```

```python
X = cases.drop(["total_cases"],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

# Modelling
## 1. Logistic Regression

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
logisticregressor = LogisticRegression()

logisticregressor.fit(X_train,Y_train)
```

```python
logisticregressor.score(X_test,Y_test)
```

#### Thus, the accuracy for our Logistic Regression model is very low - 5%

## 2. Random Forest Regressor

```python
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor
randomforestregressor = RandomForestRegressor()

randomforestregressor.fit(X_train,Y_train)
```

```python
randomforestregressor.score(X_test,Y_test)
```

#### Thus, accuracy for our Random Forest Regressor is very good - 99%
